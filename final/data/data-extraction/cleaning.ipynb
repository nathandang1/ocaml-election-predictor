```python
import pandas as pd

def process_data(fname):
    # Load the first CSV into a DataFrame
    df = pd.read_csv(fname)

    # Create an empty list to hold the rows for the second DataFrame
    rows = []

    # Get unique years
    years = df['year'].unique()
    state = df['state'].unique()[0]
    
    # Process each year
    for year in years:
        # Filter data for the current year
        df_year = df[df['year'] == year]

        # Extract the percentage votes for each party
        percent_votes_rep = df_year[df_year['party_detailed'] == 'REPUBLICAN']['percent_votes'].values[0]
        percent_votes_dem = df_year[df_year['party_detailed'] == 'DEMOCRAT']['percent_votes'].values[0]

        # Determine the winner
        winner = 'rep' if percent_votes_rep > percent_votes_dem else 'dem'

        # Create a dictionary for the row
        row = {
            'year': year,
            'state': state,
            'percent_votes_rep': percent_votes_rep,
            'percent_votes_dem': percent_votes_dem,
            'winner': winner
        }

        # Append the row to the list of rows
        rows.append(row)

    # Create the second DataFrame from the list of rows
    df_second = pd.DataFrame(rows)

    # Write the second DataFrame to a CSV file
    df_second.to_csv(f'{fname}', index=False)
```

```python
import os

directory = "state-extract"

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        # Process the file as needed
        print(file_path)
        process_data(file_path)
```
